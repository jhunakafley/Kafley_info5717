```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("C:/Care_Provider_Languages_they_speak.csv",encoding = "utf-8",index_col=False)
df.head()
```

```python
dataset=df.loc[~pd.isnull(df.Number_language)]
dataset = df[['Care_Provider_Title','Number_language']][(df['Number_language'] <=10)]

dataset_new=dataset.groupby('Number_language').count()
```

```python
ax=dataset.hist()

plt.title('Number of languages Care Providers speak')
plt.xlabel('Number_language')
plt.ylabel('Frequency')
plt.show()
```

```python
dataset_pie=dataset.groupby('Care_Provider_Title').count()


dataset_pie=dataset_pie.nlargest(8,'Number_language')

dataset_pie.plot(kind='pie', subplots=True, figsize=(8, 8),autopct='%.2f',fontsize=10)

plt.legend(loc=0, fontsize=8)


plt.show()
```

```python
dataset_new=dataset.groupby('Number_language').count()
dataset_new=dataset_new.reset_index()
plt.scatter(dataset_new.Number_language,dataset_new.Care_Provider_Title)
plt.show()
```
